## Identifying and Removing "Missing" Data

The missing data points in this dataset are refered to as the string 'missing'. In order to remove the string "missing" from the columns in this dataset you would use the code below. Through utilizing this code, any occurance of 'missing' will be dropped within each column specified. In order to ensure all the missing elements are removed from this dataset, this code will be run for each column.

```python
df = df.drop(df[df['State of Residence'] == 'missing'].index)
df
```

```python
df = df.drop(df[df['Patient Age'] == 'missing'].index)
df
```

```python
df = df.drop(df[df['Visited Last Month'] == 'missing'].index)
df
```

```python
df = df.drop(df[df['Average BP'] == 'missing'].index)
df
```

```python
df = df.drop(df[df['Average Heart Rate'] == 'missing'].index)
df
```

```python
df = df.drop(df[df['Last Visit (days ago)'] == 'missing'].index)
df
```

##Removing Duplicate Rows and Columns

In order to remove duplicate data found in this dataset, the code below is utilized. After running this code we can see that after the removale of 'missing' data, there are no duplicate enteries found.

```python
df.drop_duplicates(inplace=True)
df
```

##Cleaning Column Names

In order for the column titles to be compatible while using Python to clean this dataset, next we will be clearing the column names of white spaces and special characters using the code below.

```python
import re
def cleaning_columns (df):
    def clean_name(name):
        cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
        return cleaned_name.lower()

    df.columns = [clean_name(col) for col in df.columns]
    return df

df = cleaning_columns(df)
df
```
